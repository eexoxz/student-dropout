```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
# Load CSV file
df = pd.read_csv("StudentPerformanceFactors.csv")

# Display first few rows
print(df.shape)
df.head()
print (df.info())
```

```python
df.isnull().sum()
```

```python
df.fillna({
    'Teacher_Quality': df['Teacher_Quality'].mode()[0],
    'Parental_Education_Level': df['Parental_Education_Level'].mode()[0],
    'Distance_from_Home': df['Distance_from_Home'].mode()[0]
}, inplace=True)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cat_cols = df.select_dtypes(include=['object']).columns
print(" Columns to encode:", list(cat_cols))
for col in cat_cols:
    df[col] = le.fit_transform(df[col])
    print(f"  Encoded column: {col}")
print("\n Dataset Info After Encoding:")
print(df.info())
display(df.head())
```

```python
from sklearn.preprocessing import MinMaxScaler
scaled_df = df.copy()
cols_to_scale = df.columns.drop('Exam_Score')

scaler = MinMaxScaler()
scaled_df[cols_to_scale] = scaler.fit_transform(df[cols_to_scale])

print(" Scaling complete! All features are now between 0 and 1.")
display(scaled_df.head())
```

```python
constant_cols = [col for col in df.columns if df[col].nunique() == 1]
print("âš ï¸ Constant columns (no variation):", constant_cols)
duplicate_cols = df.columns[df.T.duplicated()].tolist()
print("âš ï¸ Duplicate columns:", duplicate_cols)
irrelevant_cols = [col for col in df.columns if 'id' in col.lower() or 'unnamed' in col.lower()]
print("âš ï¸ Possibly irrelevant columns:", irrelevant_cols)
print("\nðŸ”Ž Columns with missing values (if any):")
print(df.isnull().sum()[df.isnull().sum() > 0])
print("\nâœ… Summary:")
print("Total columns:", len(df.columns))
print("Column names:", list(df.columns))
```

```python
scaled_df.to_csv("student_performance_cleaned.csv", index=False)
print("ðŸ’¾ Final cleaned and scaled dataset saved as student_performance_cleaned.csv")
```
